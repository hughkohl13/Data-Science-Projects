```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import  pandas as pd
```

```python
df = pd.read_csv('draft_ranks_final.csv')
```

```python
df.head()
```

```python
my_url = 'https://www.nfldraftgeek.com/2020-nfl-mock-draft-2-0-fredrickson/'
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html, "html.parser")
```

```python
draft_geek = page_soup.find_all('li')
draft_geek = draft_geek[25:57]
players_ = []
players_clean = []
for i in draft_geek:
    players_.append(i.text)

for i in players_:
    split_string = i.split(": ", 1)
    players_clean.append(split_string[1])
```

```python
df_draftgeek = pd.DataFrame([sub.split(",") for sub in players_clean])
df_draftgeek.shape
```

```python
df_draftgeek.columns = ['player', 'position', 'school']
df_draftgeek['draftgeek'] = df_draftgeek.index + 1
```

```python
df_draftgeek.drop(['position', 'school'], axis=1, inplace=True)
```

```python
df = df.merge(df_draftgeek, left_on='player', right_on='player', how='outer')
df.shape
```

```python
my_url = 'https://www.cbssports.com/nfl/draft/news/three-round-2020-nfl-mock-draft-chargers-pass-on-qb-justin-herbert-jordan-love-fall-to-bottom-of-1st-round/'
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html, "html.parser")
```

```python
cbs = page_soup.find_all('span', {'class':'PlayerName'})
players_ = []
players_clean = []
for i in cbs:
    players_.append(i.text)

for i in players_:
    split_string = i.split("\n", 3)
    players_clean.append(split_string[2][24:])  

while("" in players_clean) : 
    players_clean.remove("")
    
df_cbs = pd.DataFrame([sub.split(",") for sub in players_clean])
df_cbs['cbs_wilson'] = df_cbs.index + 1
df_cbs.columns = ['player', 'cbs_wilson']
df_cbs.head()
```

```python
df = df.merge(df_cbs, left_on='player', right_on='player', how='outer')
```

```python
df.to_csv('draft_ranks_adjusted.csv')
```
