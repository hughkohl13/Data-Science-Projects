```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import  pandas as pd
```

```python
# chad reuter puyll from nfl network
urllist = ['http://www.nfl.com/news/story/0ap3000001109774/article/chad-reuter-sevenround-2020-nfl-mock-draft-round-1',
          'http://www.nfl.com/news/story/0ap3000001109775/article/chad-reuter-sevenround-2020-nfl-mock-draft-round-2',
          'http://www.nfl.com/news/story/0ap3000001109776/article/chad-reuter-sevenround-2020-nfl-mock-draft-round-3'
          ]

players_clean = []

def first_scrape(lst):
    
    for i in lst:
        my_url =  i
        uClient = uReq(my_url)
        page_html = uClient.read()
        uClient.close()
        
        page_soup = soup(page_html, "html.parser")
        
        draft_ = page_soup.find_all("span", {"class": "team-name"})
        
        players_ = []

        for i in draft_:
            players_.append(i.text)
        
        for i in players_:
            split_string = i.split("-", 1)
            players_clean.append(split_string[0][:-1])
            
    print(players_clean)

first_scrape(urllist)
```

```python
players_clean[:5]
```

```python
df = pd.read_csv('draft_ranks.csv') 
df
```

```python
df_reuter = pd.DataFrame(players_clean)
df_reuter['reuter'] = df_reuter.index + 1
df_reuter.columns = ['player', 'reuter']
df_reuter
```

```python
df = df.merge(df_reuter, left_on='player', right_on='player', how='outer')
df.shape
```

```python
#bleacher report top 100
my_url =  'https://bleacherreport.com/articles/2868550-2020-nfl-draft-big-board-matt-millers-latest-rankings'
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
        
page_soup = soup(page_html, "html.parser")
```

```python
bleacher_lst = page_soup.find_all('span')[15:94]
```

```python
bleacher_plyers = []

for i in bleacher_lst:
    bleacher_plyers.append(i.text)
```

```python
df_bleacher = pd.DataFrame([sub.split(",") for sub in bleacher_plyers])
df_bleacher
```

```python
df_bleacher.columns = ['player', 'position', 'school']
```

```python
import re 
  
def remove(lst): 
    pattern = '[0-9]'
    list = [re.sub(pattern, '', i) for i in lst] 
    return list
  
# Driver code  
lst = df_bleacher['player'] 
print(remove(lst)) 
df_bleacher['player'] = remove(lst)
df_bleacher.head()
```

```python
bl_lst = []
for i in df_bleacher['player']:
    i = i[2:]
    bl_lst.append(i)
    
df_bleacher['player'] = bl_lst
df_bleacher.head()
```

```python
df_bleacher['bleacher'] = df_bleacher.index + 1
df_bleacher.drop(['position', 'school'],axis=1, inplace=True)
df_bleacher.head()
```

```python
df = df.merge(df_bleacher, left_on='player', right_on='player', how='outer')
```

```python
# second SI scrape
my_url =  'https://www.si.com/nfl/2020/04/19/draft-big-board-updated-top-255-prospects'
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html, "html.parser")
```

```python
first_50 = page_soup.find_all("strong")
first_50 = first_50[:50]
last_50 = page_soup.find_all("p")
last_50 = last_50[107:157]
top_100 = first_50+last_50
```

```python
top_100players = []

for i in top_100:
    top_100players.append(i.text)

df_si2 = pd.DataFrame([sub.split(",") for sub in top_100players])
df_si2.columns = ["player", "position", "school", "height", "weight"]
```

```python
si2_lst = []

def remove_10(player):
    for i in player:
        i = i[3:]
        si2_lst.append(i)
    
def remove_99(player):
    for i in player:
        i = i[4:]
        si2_lst.append(i)

    
lst1 = df_si2['player'].iloc[0:9]
lst2 = df_si2['player'].iloc[9:99]
lst3 = df_si2['player'].iloc[99]
remove_10(lst1)
remove_99(lst2)
si2_lst.append('Davon Hamilton')
df_si2['player'] = si2_lst
df_si2['si2_rank'] = df_si2.index + 1
df_si2.drop(['school', 'height', 'weight', 'position'],axis=1, inplace=True)
df_si2.head()
```

```python
df = df.merge(df_si2, left_on='player', right_on='player', how='outer')
df.shape
```
