```python
urllist = ['JacobK',
          'FATSAK',
          'GoPats1287',
          'JustinB',
          'Saucey19',
          'Pouhy',
           'Logantooreal',
           'rayt',
           'username2003',
           'VFarm1990',
           'bkru3813',
           'emt6000',
           'Jackd1305',
           'Kamikaze_Justice',
          ]

def add_url(urllst): 
    ds_lst2 = []
    for i in urllst:
        my_url =  'https://www.draftsite.com/mydraftsite/' + i + '/nfl/user-player-rankings/2020/'
        uClient = uReq(my_url)
        page_html = uClient.read()
        uClient.close()
        
        page_soup = soup(page_html, "html.parser")

        ds_ = page_soup.find_all("p", {"class": "mb-0"})
        ds_lst = []
        
        def ds_rank(lst):
            for item in lst:
                ds_lst.append(item.text)
        ds_rank(ds_)
          
        ds_lst = ds_lst[:503]
        ds_lst = ds_lst[4::5]
        ds_lst2.append(ds_lst)

#         ds_df = pd.DataFrame([sub.split(",") for sub in ds_lst])
#         ds_df[i] = (ds_df+1).index + 1

#         (ds_df+i).columns = ['player', i]
#         df_new = pd.merge(df_new, (ds_df+i))
    return ds_lst2

add_url(urllist)
```

```python
#remove empty values before putting into dataframe
for i in ds_lst2:
    while("" in i): 
        i.remove("") 
```

```python
# turn list of lists in ds_lst2 to list of dataframes

new_dfs = []

for i in ds_lst2:
    df_ = pd.DataFrame([sub.split(",") for sub in i])
    df_['rank'] = df_.index + 1
        
    df_.columns = ['player', 'rank']
    new_dfs.append(df_)
```
