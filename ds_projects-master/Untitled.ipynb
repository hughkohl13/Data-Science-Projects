```python
from bs4 import BeautifulSoup
import requests
import  pandas as pd
import numpy as np
```

```python
job_roles = ['solutions+engineer',
            'solutions+architect',
            'sales+engineer',
            'pre-sales',
            ]

url_test = 'https://www.simplyhired.com/search?q=solutions+engineer&l=San+Francisco%2C+CA'

scrapes = requests.get(job_roles)

soup = BeautifulSoup(scrapes.text, 'html.parser')

role_card = soup.find_all('div', {'class':"SerpJob-jobCard card"})

urls = []

for i in role_card:
    url = 'https://simplyhired.com' + i.find('a', {'class':'card-link'}, href=True)['href']
    urls.extend([url])
    
scrape = []
for u in urls:
    scrape_requests = requests.get(u)
    second_scrape = BeautifulSoup(scrape_requests.text, 'html.parser')
    role = second_scrape.find("h1").text
    company = second_scrape.find('span', {'class':'company'}).text
    description = second_scrape.find('div', {'class':'viewjob-description ViewJob-description'}).text
    scrape.extend([[role,company,description]])

scrape
```

```python
city_db = pd.read_csv('./uscities.csv')
city_db.head()
```

```python
def clean_city(city_list):
    city_list['query_string'] = city_list['city'].str.replace(" ", "+")
    city_list['query_string'] = city_list['query_string'] + '%2C+'
    city_list['query_string'] = city_list['query_string'] + city_db['state'] 
    return city_list
    
clean_city(city_db)
```
